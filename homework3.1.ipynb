```
from numpy import arange, empty
from math import cos, sin, pi, sqrt
from pylab import plot, show, xlabel, ylabel, title,imshow
```

```
def J(m, i):   
    N = 1000
    a = 0
    b = pi
    h = (b - a) / N
    oddsum = 0
    evensum = 0
    for k in range(1, N, 2):
        oddsum += cos(m * (a + k * h) - i * sin((a + k * h)))
    for k in range(2, N, 2):
        evensum += cos(m * (a + k * h) - i * sin((a + k * h)))
    return 1 / pi * 1 / 3 * h * (cos(m * a - i * sin(a)) + cos(m * b - i * sin(b)) + 4 * oddsum + 2 * evensum)

x = linspace(0,20,100) 
J0 = []
J1 = []
J2 = []
for j in x:
    J0.append(J(0, j))
    J1.append(J(1, j))
    J2.append(J(2, j))
plot(x, J0)
plot(x, J1)
plot(x, J2)
xlabel("x")
ylabel("J")
title("Bessel Function")
show()

def I(x, y):
    if sqrt(x**2 + y**2) == 0:
        return 1/4
    kr = 2 * pi / 0.5 * sqrt(x**2 + y**2)
    return (J(1, kr)/ kr)**2

intensity = empty([200, 200], float)

for i in range(200):
    y = 0.01 * i
    for j in range(200):
        x = 0.01 * j
        intensity[i, j] = I(x - 1 , y - 1 )
imshow(intensity, vmax=0.005)
xlabel("x(10^(-8)m)")
ylabel("y(10^(-8)m)")
title("Intensity")
show()

```
